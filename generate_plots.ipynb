```python
import pandas as pd
%matplotlib inline

filepaths = ['unet_history.csv', 'deeplabv3_history.csv', 'unet_res_history.csv', 'unet_invres_history.csv']


def plot_history(filepath):
    history_df = pd.read_csv(filepath)
    history_df[['loss', 'val_loss']].plot()
    history_df[['dice_coef', 'val_dice_coef']].plot()
    
for file in filepaths:
    plot_history(file)
```
