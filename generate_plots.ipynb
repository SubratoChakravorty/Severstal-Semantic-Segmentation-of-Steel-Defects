```python
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline

filepaths = ['base_UNET_history_updated.csv', 'deeplabv3_history.csv', 'UNET_MCAD_history.csv', 'UNET_MCAD_invres_history.csv']

# Model names
titles = ['Base U-net', 'deeplabv3+', 'Unet-MCAD', 'UNET_MCAD_invres']

df = pd.DataFrame()

# Plot parameters
params = ['loss','val_loss','dice_coef', 'val_dice_coef']

# Create a dataframe with all the parameters for all models
for i, filepath in enumerate(filepaths):
    history_df = pd.read_csv(filepath)
    title = titles[i]
    df[title+'_loss'] = history_df['loss']
    df[title+'_val_loss'] = history_df['val_loss']
    df[title+'_dice_coef'] = history_df['dice_coef']
    df[title+'_val_dice_coef'] = history_df['val_dice_coef']

# Plot a model for each parameter
for param in params:
    plot_list = [title+'_'+param for title in titles]
    df[plot_list].plot(title=param)
    
    



```
